```python
import matplotlib.pyplot as plt
import numpy as np
import prosail
from scipy.signal import resample

# Given data
result = prosail.run_prosail(n=1.5, cab=40, car=8, cbrown=0.0, cw=0.01, cm=0.009, lai=3.0, lidfa=-0.35, lidfb=-0.15, rsoil=0.2, psoil=0.3, hspot=0.01, tts=30.0, tto=10.0, psi=0.0, typelidf=1)

# Wavelength range (400 to 2500 nm)
wavelengths = np.linspace(400, 2500, result.shape[0])


# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(wavelengths, result, label='Reflectance Spectrum')
plt.title('Reflectance Spectrum (400-2500 nm)')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
